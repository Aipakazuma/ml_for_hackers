```python
%%R
# 平均値と中央値の違い

# 今回使用するベクトル
vec =  c(1, 2, 3, 4, 5, 6)

# 平均値を求める関数
my.mean <- function(x) {
    return(sum(x) / length(x))
}
my.mean(vec) # 3
```

```python
%%R

# でもRにはmean()って関数がすでに用意されている

mean(vec) # 3
```

```python
%%R
# 中央値を求める関数
my.median <- function(x) {
    # 中央値は真ん中の値を取るので、まずはソートする
    sorted.x <- sort(x)
    
    # 偶数か、そうでないかで計算が変わる
    # 偶数の場合は、例えばc(1, 2, 3, 4)だった場合、中央となるべき値は2つ（2, 3）であり、その平均を出す
    # 奇数であれば、例えば c(1, 2, 3)だった場合、真ん中の2となる
    if (length(x) %% 2 == 0) {
        indices <- c(length(x) / 2, length(x) / 2 + 1)
        return(mean(sorted.x[indices]))
    } else {
        index <- ceiling(length(x) / 2)
        return(sorted.x[index])
    }
}

my.median(vec) # 3.5
```

```python
%%R

# meanと同様、中央値もRは関数が用意されている
median(vec) # 3.5
```

```python
%%R

# 平均値からどれだけ離れているかを出力する関数を用意
my.var <- function(x) {
    m <- mean(x)
    return(sum((x - m) ^ 2) / length(x))
}

my.var(heights)
```

```python
%%R

# これもRに(ry

var(heights)
```

```python
%%R
# 比較してみる
var(heights) - my.var(heights) #  0.001480347 <- あり？
```

```python
%%R

# 説明をみたけどよくわからないので、ざっくりと
# Rの組み込み版は計算がちょっとちがくて、n / (n -1)という係数をかけている
# それが一般的な公式らしい

my.var <- function(x) {
    m <- mean(x)
    return(sum((x - m) ^ 2) / (length(x) - 1))
}
my.var(heights) - var(heights) # 0
```

なぜなら他にもベル型の分布が存在するからです〜。

んで、最頻値の話に入ります。

グラフを出力するコードがないので、適当にサマリを残していく。

* 最頻値は密度プロットを生成した場合、ベルのピークになる。あと最頻値は密度プロットのグラフがわかりやすい
* 最頻値が定義できたぞ。ここで正規分布の性質のひとつが最頻値を１つ持っていて、平均値と中央値が同じなんだって。わけわかめ。
* １つの最頻値をもつ分布は単峰型、２つは双峰型、２つ以上の最頻値を保つ場合は多峰型である。
* もうひとつ重要な性質はデータが対象か歪んでいる（非対称）。うん２つあんじゃね？
* 上の見方でまた分岐する。
* 正規分布は対象の形状をしている。
* 逆に非対称はガンマ分布と呼ばれていて右方向に歪んでいる。
* 特性の最後は薄い裾と熱い裾である。もう何言っているかわかんね。


```python
%%R

# やっとグラフが打てる(´；ω；｀)ﾌﾞﾜｯ
set.seed(1)
normal.values <- rnorm(250, 0, 1)
cauchy.values <- rcauchy(250, 0, 1)

range(normal.values)
range(cauchy.values)
```

```python
%%R

# 厚い裾　コーシー分布
ggplot(data.frame(X = normal.values), aes(x = X)) + geom_density()
```

```python
%%R

# 薄い裾
ggplot(data.frame(X = cauchy.values), aes(x = X)) + geom_density()
```

薄い裾の分布は、例えば99%の値が平均値からあまり遠くないとこおろに位置している。
例えば、正規分布において99%の値が標準偏差の3倍までの距離いないに含まれるようなケースです。

？

コーシー分布（厚い布）は標準偏差の3倍までの距離以内に値の90%しか含まれない。
さらに平均値から離れるに連れ、2つのタイプの分布の違いはより大きくなります。正規分布の場合は標準偏差の6倍以上の部分にはほぼ値が存在しないけど、コーシー分布の場合は5%ほど値がそれ以上遠くに存在しているよね？


?

なんとなくだけどわかった。（多分）




つまりつまり、正規分布の特性のおさらいをしよう

### 正規分布の特性

* 単峰分布
* 対象分布
* 薄い裾のベル型

### コーシー分布の特性

* 単峰型（？）
* 対象
* 厚い裾のベル型


グラフはおかしいと思う（確信）

でも、データはすべて使っている。ので、逆手の手法を取ることができます。

予想しようとする値が数値の場合、ある列を利用して別の列の値を予測することを「回帰」と呼ぶ。覚えた。

それに対しラベルを予測する場合、それは「分類」と呼ばれる。


今度は性別で色分けてみる。

```python
%%R

ggplot(heights.weights, aes(x = Height, y = Weight, color = Gender)) + geom_point()
```

性別を分けたら分類することができた。これが「分類」のゴールじゃないかな。
つまり与えられた身長と体重からその人の性別を判別することができるはず。
=> 他の情報から性別を推定している。


分離超平面っていう洒落た名前が付いている線がある。（ちなみに英語だとseparating hyperplane）

データを２つに分断しているので「分離」超平面というわけ。
見方としては、片側では女性であることが推定でき、逆側では男性であると推定できる。

データセットを使った結果としては、92%の正答率となる。

```python
%%R

heights.weights <- transform(heights.weights, Male = ifelse(Gender == 'Male', 1, 0)) # Mailだったら -> 1、女性だったら 0
logit.model <- glm(Male ~ Height + Weight, data = heights.weights, family = binomial(link = 'logit'))

ggplot(heights.weights, aes(x = Weight, y = Height, color = Gender)) + # さっきとはwidthとheightが逆
  geom_point() + 
  geom_abline(intercept =  - coef(logit.model)[1] / coef(logit.model)[2],
              slope = - coef(logit.model)[3] / coef(logit.model)[2],
              color = 'black') +
  coord_flip() +
  theme(axis.text.x = element_text(angle = 0, hjust = 1)) # 横軸ラベルを横に

#  stat_abline(intercept = - coef(logit.model)[1] / coef(logit.model)[2],
#    slope = - coef(logit.model)[3] / coef(logit.model)[2],
#   geom = 'abline',
#    color = 'black')
```
