## 日付文字列を変換し、正しくないデータも処理する

正しくないデータを探す。まずはこれ
使いたい値は「YYYYMMDD」の値なので、それ以外を探せば良い。

`nchar(ufo$DateOccurred) != 8` これで8文字以外のデータを探す。第２引数で条件でTRUEだった値を返す。第３引数で（ｒｙ

んで、どんなけ存在するから`length()`を使って確認する。
最後に必要データを取得する。（上書きなことに注意）

```python
%%R
ufo$DateOccurred <- as.Date(ufo$DateOccurred, format = "%Y%m%d")
ufo$DateReported <- as.Date(ufo$DateReported, format = "%Y%m%d")
```

`DateOccurred` と `DateReported` のデータを取得

```python
%%R
bool.return <- function(bool) {
  return(bool)
}

bool.return(FALSE)
```

```python
%%R
get.location <- function(l) {
  split.location <- tryCatch(strsplit(l, ",")[[1]],                  # try catchしているのかな。よくわからん
                             error = function(e) return(c(NA, NA)))  # error時はこう返せってことかなよくわからん
  clean.location <- gsub("^ ","",split.location)                    # これはなんだ。。。？あれか？先頭に空文字が存在したら置換ってことか。
  if (length(clean.location) > 2) { # 置換した値が入っているはず。配列が3以上であれば、NA配列で返す
    return(c(NA,NA))                       # なぜ３つかって？データを見てみよう。（アメリカは都市名,州名なので２つになるはず）
  } else {
    return(clean.location)
  }
}
```

```python
%%R
test <- "20,20"
strsplit(test, ",")
```

おおー、これでsplitできるのね。

```python
%%R
test <- "20 20"
strsplit(test, ",")
```

あれ？文字が存在しない場合はエラーになるらしいけど、、、なんでだろー？

```python
%%R
test <- "test"
strsplit(test, "")
```

おおー。これで一文字ずつの配列ができた。

```python
%%R
test2 <- " 20"
gsub("^ ", "", test2)
```

```python
%%R
ufo.us$YearMonth <- strftime(ufo.us$DateOccurred, format = "%Y-%m")
```

```python
%%R
sightings.counts <- ddply(ufo.us, .(USState,YearMonth), nrow)
head(sightings.counts)
```

↑のように、年月単位でUFO目撃件数を取得できた。
しかし、akは1990年2月のデータがないため、これではグラフが作成できない（らしい）

```python
%%R
date.range <- seq.Date(from = as.Date(min(ufo.us$DateOccurred)),
                       to = as.Date(max(ufo.us$DateOccurred)),
                       by = "month")
head(date.range)
```

```python
%%R
date.strings <- strftime(date.range, "%Y-%m")
head(date.strings)
```

```python
%%R
states.dates <- lapply(state.abb, function(s) cbind(s, date.strings) )
head(states.dates, 2)
```

```python
%%R
states.dates <- data.frame(do.call(rbind, states.dates),
                           stringsAsFactors = FALSE)
head(states.dates)
```

```python
%%R
all.sightings <- merge(states.dates,
                       sightings.counts,
                       by.x = c("s", "date.strings"),
                       by.y = c("USState", "YearMonth"),
                       all = TRUE)
```

```python
%%R
names(all.sightings) <- c("State", "YearMonth", "Sightings")
```

```python
%%R
all.sightings$Sightings[is.na(all.sightings$Sightings)] <- 0
```

```python
%%R
all.sightings$YearMonth <- as.Date(rep(date.range, length(state.abb)))
```

```python
%%R
all.sightings$State <- as.factor(all.sightings$State)
```

```python
%%R
state.plot <- ggplot(all.sightings, aes(x = YearMonth,y = Sightings)) +
  geom_line(aes(color = "darkblue")) +
  facet_wrap(~State, nrow = 10, ncol = 5) + 
  theme_bw() + 
  scale_color_manual(values = c("darkblue" = "darkblue"), guide = "none") +
  scale_x_date(date_breaks = "5 years", date_labels =  '%Y') +
  xlab("Years") +
  ylab("Number of Sightings") +
  ggtitle("Number of UFO sightings by Month-Year and U.S. State (1990-2010)")
```

```python
%%R
print(state.plot)
```

```python
%%R
state.pop <- read.csv(file.path('census.csv'), stringsAsFactors=FALSE)
```

```python
%%R
state.pop$abbs <- sapply(state.pop$State, function(x) state.abb[grep(paste('^', x, sep=''), state.name)])
all.sightings$Sightings.Norm <- sapply(1:nrow(all.sightings), 
    function(i) all.sightings$Sightings[i] / state.pop$X2000[which(state.pop$abbs== all.sightings$State[i])])
```

```python
%%R
state.plot.norm <- ggplot(all.sightings, aes(x = YearMonth,y = Sightings.Norm)) +
  geom_line(aes(color = "darkblue")) +
  facet_wrap(~State, nrow = 10, ncol = 5) + 
  theme_bw() + 
  scale_color_manual(values = c("darkblue" = "darkblue"), guide = "none") +
  scale_x_date(date_breaks = "5 years", date_labels = '%Y') +
  xlab("Years") +
  ylab("Per Capita Number of Sightings (2000 Census)") +
  ggtitle("Number of UFO sightings by Month-Year and U.S. State (1990-2010)")
```
